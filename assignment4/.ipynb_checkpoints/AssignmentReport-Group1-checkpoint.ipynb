## task 1b)
A true positive is a a correct detection. This means that the detection has a IOU ≥ threshold.
A false positvie is a wrong detection. This means that the detection has a IOU < threshold.
$$Precision = \frac{True \quad positive}{True \quad positive + false\quad positive}$$
$$Recall = \frac{True \quad positive}{True \quad positive + false\quad negative}$$
